```python
#Import Necessary Libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset: Download the Iris dataset from the provided link and load it into a Pandas DataFrame:
df = pd.read_csv('Iris.csv')  # Adjust the file path if needed
```

```python
#Inspect the Dataset:
df.info()
```

```python
#Inspect data distribution with df.describe().
df.describe()
```

```python
#understand the dataset structure.
df.head()
```

```python
#Exploratory Data Visualization
#Understand Data Characteristics:
#Check for missing values using:
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```

```python
#Correlation Matrix:
#Use Seaborn to visualize correlations between features:
# Filter only numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['number'])

# Check if there are numeric columns
if numeric_df.empty:
    print("No numeric columns available for correlation.")
else:
    # Visualize the correlation matrix
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

```

```python
#### Answering Assignment Questions
# 1. What is the distribution of sepal lengths and widths among the Iris species?
#Visualization Used: Box plots.
sns.boxplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Distribution of Sepal Lengths Among Iris Species')
plt.show()

sns.boxplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Distribution of Sepal Widths Among Iris Species')
plt.show()
```

```python
#Inference:
#Sepal Length: The distribution shows that Iris Virginica generally has the longest sepals, followed by Versicolor and then Setosa. There is minimal overlap in the ranges of sepal length between species, making it somewhat useful for classification.
#Sepal Width: Setosa has the widest sepals on average, while Virginica and Versicolor have narrower, overlapping ranges.

#Story:
#The sepal measurements give us the first glimpse of differentiation among species. Iris Setosa stands out with its broad sepals, while the lengths hint at a progression from Setosa to Virginica. This initial observation suggests sepal characteristics could be used to identify certain species but may not be sufficient on their own.
```

```python
# 2. How does the petal length vary across different Iris species?
#Visualization Used: Box plot.
sns.boxplot(x='Species', y='PetalLengthCm', data=df)
plt.title('Variation of Petal Length Among Iris Species')
plt.show()
```

```python
#Inference:
#Petal length is a highly distinguishing feature.
#Setosa has significantly shorter petals, with little overlap between its range and the ranges for Versicolor and Virginica.
#Virginica has the longest petals, with some overlap with Versicolor.

#Story:
#Petal length reveals a striking difference among species, acting as a primary separator for Setosa versus the other two species. While Virginica and Versicolor show some overlap, the clear separation of Setosa hints at how the species might have adapted differently, possibly for pollination by specific insects.
```

```python
# 3. Is there a relationship between sepal length and petal length?
#Visualization Used: Scatter plot.
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=df)
plt.title('Relationship Between Sepal Length and Petal Length')
plt.show()
```

```python
#Inference:
#There is a strong positive correlation between sepal length and petal length.
#This relationship is most pronounced in Versicolor and Virginica. However, Setosa forms a distinct cluster with much smaller petal lengths, regardless of sepal size.

#Story:
#The relationship between sepal length and petal length shows how these dimensions scale together, especially in Versicolor and Virginica. This could indicate that larger sepals and petals evolve in tandem, possibly to attract pollinators or accommodate specific environmental conditions. Setosa, with its smaller petals, stands out as an exception, likely reflecting a unique ecological niche.
```

```python
# 4. How does the petal width vary across different Iris species?
#Visualization Used: Box plot.
sns.boxplot(x='Species', y='PetalWidthCm', data=df)
plt.title('Variation of Petal Width Among Iris Species')
plt.show()
```

```python
#Inference:
#Petal width mirrors the pattern observed in petal length. Setosa has the narrowest petals, while Virginica has the widest, with Versicolor in between.
#There is minimal overlap in petal width between Setosa and the other species, but Virginica and Versicolor share a similar range.

#Story:
#Petal width complements the insights from petal length, reinforcing the clear distinction of Setosa from the other two species. The overlapping ranges between Versicolor and Virginica suggest their petals serve similar functions, potentially attracting similar pollinators or thriving in similar environments.

```

```python
# 5. Are there any patterns observable in the pairwise relationships between features?
#Visualization Used: Pair plot.
sns.pairplot(df, hue='Species')
plt.show()
```

```python
#Inference:
#Pairwise plots reveal clear clustering of species, particularly with petal measurements.
#Setosa forms a completely distinct cluster, while Versicolor and Virginica show some overlap in certain dimensions.
#Petal length and petal width together provide the strongest separation among species.

#Story:
#The pairwise relationships highlight how the species can be distinguished when multiple features are combined. Setosa is the most unique, forming an isolated cluster. Versicolor and Virginica show a closer evolutionary relationship, as evidenced by their overlapping clusters in some dimensions. These patterns hint at the evolutionary divergence of the species and their adaptation to different ecological roles.
```

```python
# 6. Can we distinguish between Iris species based on sepal and petal measurements?
#Visualization Used: Scatter plot with hue as species.
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm', hue='Species', data=df)
plt.title('Petal Length vs. Petal Width by Species')
plt.show()

sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm', hue='Species', data=df)
plt.title('Sepal Length vs. Sepal Width by Species')
plt.show()
```

```python
#Inference:
#Petal measurements provide near-perfect separation between species.
#Sepal measurements also show some separation, but there is noticeable overlap between Versicolor and Virginica.
#Combining both petal and sepal dimensions yields a strong basis for distinguishing species.

#Story:
#The sepal and petal dimensions together tell a compelling story of species differentiation. While sepals provide general trends, petals reveal the fine-grained details needed for classification. The clear separation of Setosa underscores its unique evolutionary pathway, while the partial overlap of Versicolor and Virginica points to their closer ecological or evolutionary connections. This visualization confirms that the Iris dataset's features are well-suited for distinguishing species with high accuracy.
```
