```python
# DATA VISUALIZATION on two datasets the Iris dataset and the Diabetes dataset using Matplotlib and Seaborn libraries.
#1. Setup and Data Loading
# Install necessary libraries (if not installed)
!pip install seaborn pandas matplotlib

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
#iris_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/iris.csv"
iris_df = pd.read_csv('Iris.csv')

# Load the Diabetes dataset
#diabetes_url = "https://raw.githubusercontent.com/mathchi/diabetes-data-set/master/diabetes.csv"
diabetes_df = pd.read_csv('diabetes.csv')

# Display the first few rows of each dataset
print("Iris Dataset")
print(iris_df.head())

print("\nDiabetes Dataset")
print(diabetes_df.head())
```

```python
# 2. Exploratory Data Analysis (EDA)
# 2.1. Iris Dataset Visualizations
# 1. Distribution of Sepal Lengths and Widths Among the Iris Species
# We'll create box plots to understand the distribution of SepalLengthCm and SepalWidthCm for each species.

# Distribution of Sepal Length and Width among Iris species
plt.figure(figsize=(12, 6))

# Sepal Length Distribution
plt.subplot(1, 2, 1)
sns.boxplot(x='Species', y='SepalLengthCm', data=iris_df)
plt.title('Sepal Length Distribution by Species')

# Sepal Width Distribution
plt.subplot(1, 2, 2)
sns.boxplot(x='Species', y='SepalWidthCm', data=iris_df)
plt.title('Sepal Width Distribution by Species')

plt.tight_layout()
plt.show()
```

```python
# 2. How Does Petal Length Vary Across Different Iris Species?
# We will create a box plot to see the variation of PetalLengthCm across species.

# Petal Length Distribution across Iris Species
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='PetalLengthCm', data=iris_df)
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
# 3. Relationship Between Sepal Length and Petal Length
# A scatter plot will help visualize the relationship between SepalLengthCm and PetalLengthCm.

# Relationship between Sepal Length and Petal Length
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=iris_df, palette="Set1")
plt.title('Sepal Length vs Petal Length')
plt.show()
```

```python
# 4. Pairwise Relationships Between Features
# We will use pair plot to show pairwise relationships between SepalLengthCm, SepalWidthCm, PetalLengthCm, and PetalWidthCm.

# Pairwise relationships between features
sns.pairplot(iris_df, hue='Species', palette="Set1")
plt.title('Pairwise Relationships Among Features')
plt.show()
```

```python
# 5. Can We Distinguish Between Iris Species Based on Sepal and Petal Measurements?
# A scatter plot matrix (pairplot) can show how well the species can be distinguished based on their features.

# Pairplot for distinguishing species based on all measurements
sns.pairplot(iris_df, hue='Species')
plt.title('Distinguishing Iris Species Based on Sepal and Petal Measurements')
plt.show()
```

```python
# 2.2. Diabetes Dataset Visualizations
# 6. Distribution of Age and BMI Among Diabetic Patients
# We will use histograms to show the distributions of Age and BMI.

# Distribution of Age and BMI among diabetic patients
plt.figure(figsize=(12, 6))

# Age Distribution
plt.subplot(1, 2, 1)
sns.histplot(diabetes_df['Age'], kde=True, color='skyblue')
plt.title('Age Distribution')

# BMI Distribution
plt.subplot(1, 2, 2)
sns.histplot(diabetes_df['BMI'], kde=True, color='lightgreen')
plt.title('BMI Distribution')

plt.tight_layout()
plt.show()
```

```python
# 7. How Do Blood Pressure Levels Vary Across Different Age Groups?
# A box plot will show the variation of BloodPressure across different Age groups.

# Blood Pressure variation across Age groups
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', y='BloodPressure', data=diabetes_df)
plt.title('Blood Pressure Variation Across Age Groups')
plt.show()
```

```python
# 8. Is There a Correlation Between BMI and Glucose Levels?
# A scatter plot will help us see if theres a correlation between BMI and Glucose levels.
# Correlation between BMI and Glucose

plt.figure(figsize=(8, 6))
sns.scatterplot(x='BMI', y='Glucose', data=diabetes_df, color='orange')
plt.title('Correlation Between BMI and Glucose Levels')
plt.show()
```

```python
# 9. Observable Patterns in the Relationship Between Diabetes Progression and Age
#We'll use a line plot to examine if there's a relationship between DiabetesPedigreeFunction and Age.

# Diabetes progression vs Age
plt.figure(figsize=(8, 6))
sns.lineplot(x='Age', y='DiabetesPedigreeFunction', data=diabetes_df)
plt.title('Diabetes Progression vs Age')
plt.show()
```

```python
# 10. Variation of Insulin Levels Across Different Age Groups
# We will use a box plot to see how Insulin levels vary across different Age groups.

# Insulin variation across different Age groups
plt.figure(figsize=(8, 6))
sns.boxplot(x='Age', y='Insulin', data=diabetes_df)
plt.title('Insulin Levels Across Different Age Groups')
plt.show()
```
