```python
#Step 3: Summarize the Conversation and Extract Insights (Python Code)

import requests
import json

# Function to generate conversation summary and insights, without the <think> block
def generate_conversation_summary():
    url = "http://localhost:11434/api/chat"

    # Refined prompt to generate a summary without the <think> block
    payload = {
        "model": "deepseek-r1",  # Use DeepSeek-R1 model
        "messages": [
            {"role": "user", "content": "You are an expert customer service analyst. Please review the following conversation between a customer and a support agent: The customer received the wrong jacket (red instead of black) and contacted support. The support agent apologized and offered a replacement or refund with a return label. The customer was informed that the replacement would arrive within 3-5 business days and thanked the agent for the help. Please summarize the conversation and provide key insights, without including any internal reflection or reasoning ('<think>' block)."}
        ]
    }

    # Send the POST request to the Ollama server
    response = requests.post(url, json=payload, stream=True)
    full_response = ""

    # Handle streaming response
    for line in response.iter_lines():
        if line:
            json_response = json.loads(line.decode('utf-8'))
            if 'message' in json_response:
                content = json_response['message']['content']
                full_response += content

    # Clean the output to remove any unnecessary reflections
    clean_response = remove_think_block(full_response)
    return clean_response

# Function to remove <think> block from generated content
def remove_think_block(text):
    # Remove the <think> block if present
    if "<think>" in text:
        start_index = text.find("<think>") + len("<think>")  # Find where the think block starts
        end_index = text.find("</think>")  # Find where the think block ends
        cleaned_text = text[end_index + len("</think>"):].strip()  # Keep only the conversation part
        return cleaned_text
    return text

# Generate and print the conversation summary and insights
conversation_summary = generate_conversation_summary()
print("Generated Conversation Summary and Insights:\n", conversation_summary)
```
