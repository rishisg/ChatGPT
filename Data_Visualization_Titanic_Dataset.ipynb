```python
#Import Libraries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset: Download the Titanic dataset from the provided link, and load it into a Pandas DataFrame:
df = pd.read_csv('train.csv')
```

```python
#Exploratory Data Visualization
#Understand Dataset Structure:
df.info()
```

```python
df.describe()
```

```python
df.head() #to inspect the dataset
```

```python
#Visualize Dataset Characteristics:
#Histogram for Continuous Variables: Explore age, fare, etc.
#Correlation Matrix: Visualize correlations between numerical features.

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Correlation Matrix: Visualize correlations between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

```

```python
#Identify Missing Values:
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
```

```python
#Answering Assignment Questions
#1. Distribution of Passenger Ages:
#Use a histogram or KDE plot to visualize the age distribution.
sns.histplot(df['Age'].dropna(), kde=True, bins=30)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.show()
```

```python
#Inference:
#The age distribution shows a right-skewed pattern, with most passengers being between 20 and 40 years old.
#There is a noticeable peak around 25-30 years, indicating a higher number of young adults on board.
#There were fewer elderly passengers (above 60), and some children below 10 years old were present.
```

```python
#2. Survival Rate by Passenger Class (Pclass):
#Use a bar plot.
sns.barplot(x='Pclass', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.show()
```

```python
#Inference:
#Survival rates were significantly higher for passengers in the 1st class compared to 2nd and 3rd classes.
#1st Class: ~63% survival rate.
#2nd Class: ~47% survival rate.
#3rd Class: ~24% survival rate.
#This highlights a disparity in survival chances, likely influenced by socio-economic status and access to lifeboats.
```

```python
# 3. Relationship Between Fare Paid and Survival:
# Use a box plot or scatter plot
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Paid vs. Survival')
plt.show()

```

```python
##Inference:
#Passengers who paid higher fares had a greater likelihood of survival.
#The median fare for survivors was noticeably higher than that for non-survivors.
#This trend suggests that passengers in higher classes (who generally paid more) had better access to safety resources.
```

```python
#4. Survival Rate by Gender:
#Use a bar plot
sns.barplot(x='Sex', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()
```

```python
##Inference:
#Female passengers had a much higher survival rate (~74%) compared to male passengers (~19%).
#This indicates that women were prioritized during rescue efforts, aligning with the "women and children first" policy.
```

```python
#5. Impact of SibSp and Parch on Survival:
#Use violin plots or swarm plots
sns.violinplot(x='SibSp', y='Survived', data=df, split=True)
plt.title('Impact of Siblings/Spouses (SibSp) on Survival')
plt.show()
sns.violinplot(x='Parch', y='Survived', data=df, split=True)
plt.title('Impact of Parents/Children (Parch) on Survival')
plt.show()
```

```python
##Inference:
#SibSp:
#Passengers with 1-2 siblings/spouses aboard had higher survival rates.
#Those traveling alone or with large families (SibSp > 3) had lower survival rates.
#Parch:
#Survival was highest for passengers with 1-3 parents/children aboard.
#Passengers with no family or large families (Parch > 3) were less likely to survive.
#This suggests that small family groups were more effective in coordinating and reaching safety.
```

```python
# 6. Differences in Survival Rates by Embarkation Port:
#Use a bar plot grouped by embarkation ports
sns.barplot(x='Embarked', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Embarkation Port')
plt.ylabel('Survival Rate')
plt.xlabel('Embarkation Port')
plt.show()
```

```python
##Inference:
#Passengers who embarked at Cherbourg (C) had the highest survival rate (~55%), followed by Southampton (S) (~33%), and Queenstown (Q) (~39%).
#The higher survival rate for Cherbourg passengers could be attributed to a larger proportion of 1st-class passengers embarking from this port.
```
